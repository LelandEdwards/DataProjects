```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Launch SFR.csv", encoding = "ISO-8859-1", skipinitialspace = True)
df.head ()
```

```python
df.info()
```

```python
plt.figure(figsize = (20,10))
df.groupby('Country').count()['Company'].plot(kind='bar',color='orange')
plt.xticks(rotation=50, fontsize = 12)
plt.ylabel("Total Number of Companies",fontsize=15) 
plt.xlabel("Country" , fontsize = 12)
plt.title("Companies per Country",fontsize=15)


plt.show()
```
