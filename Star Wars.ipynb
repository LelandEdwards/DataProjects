```python
df = pd.read_csv("wars.csv",index_col=0)
```

```python
df.head()
```

```python
df.dropna(axis='columns', inplace = True) 
```

```python
df.head()
```

```python
df.drop(['Originally released.1', 'Episodes.1'], axis = 1, inplace = True)
df.head()
```

```python
df.head(23)
```

```python
df.drop([0,9,17,22],inplace=True)
df.head()
```

```python
df.head(23)
```

```python
df.at[3,'Network']='Cartoon Network'
```

```python
df.head()
```

```python
df.groupby('Network').count()['Series']
```

```python
df['Network'].value_counts().plot(kind='pie',figsize=[10,15],autopct='%.f%%');
```

```python
plt.figure(figsize = (18,10))
df.groupby('Network').count()['Series'].plot(kind='bar',color='Blue')
plt.xticks(rotation=50, fontsize = 12)
plt.ylabel("Number of Series",fontsize=15) 
plt.xlabel("Network" , fontsize = 12)

plt.show()
```
